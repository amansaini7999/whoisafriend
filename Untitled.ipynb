```python
cat_cols = ['Person A', 'Person B', 'Interaction Type', 'Moon Phase During Interaction', 'PersonsNamecombined']
target = 'Friends'
```

```python
le_dict = {}

for col in cat_cols:
    le = LabelEncoder()
    le.fit(traindf[col].unique().tolist() + testdf[col].unique().tolist())
    traindf[col] = le.transform(traindf[[col]])
    testdf[col] = le.transform(testdf[[col]])
    
    le_dict[col] = le
```

```python
def baseliner(traindf, features, target, cv=3, metric='accuracy'):
    print("Baseliner Models\n")
    eval_dict = {}
    models = [xgb.XGBClassifier(), GradientBoostingClassifier(), LogisticRegression(), 
              RandomForestClassifier(), DecisionTreeClassifier(), AdaBoostClassifier()]
    print("Model Name \t |   CV")
    print("------------------------")

    for index, model in enumerate(models, 0):
        model_name = str(model).split("(")[0]
        eval_dict[model_name] = {}

        results = cross_val_score(model, traindf[features], traindf[target], cv=cv, scoring=metric)
        eval_dict[model_name]['cv'] = results.mean()

        print("%s \t | %.4f \t" % (model_name[:12], eval_dict[model_name]['cv']))
```

```python
target = 'Friends'
id_col = 'ID'

features = traindf.columns.tolist()
features.remove(target)
features.remove(id_col)

features, target
```

```python
baseliner(traindf, features, target)
```
